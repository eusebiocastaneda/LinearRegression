```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/module_5_auto.csv'
df = pd.read_csv(path)
```

```python
ind_vars = ['horsepower', 'curb-weight','engine-size','highway-mpg']
predictor = df[['price']]
```

```python
z = df[ind_vars]
lm = LinearRegression()
```

## Train the model

```python
lm.fit(z, predictor )
```

```python
yHat = lm.predict(z)
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

# Regression Plot

```python
sns.regplot(x = 'highway-mpg', y = 'price', data = df)
plt.show()
```

# Residual Plot

```python
sns.residplot(df['highway-mpg'], df['price'])
plt.show()
```

### Notice that this plot is not evenly distributed, and in fact, appears parabolic, this is a strong indicator that a linear regression model is not a valid model to use for this condition and should explore polynomial models

#### We will now show a distribution plot between the actual values versus the predicted values

```python
ax1 = sns.distplot(df['price'], hist= False, color = 'r', label = 'Actual Value')
sns.distplot(yHat, hist= False, color = 'b', label = 'Fitted Values', ax = ax1)
plt.show()
```
